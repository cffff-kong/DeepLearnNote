## one stage和two stage
### one stage
主要是ssd，yolo这种检测算法。他流程是先提取图像特征，然后生成锚框，用锚框直接去做分类，然后回归位置。值得注意的是，锚框大小都是统一的，比如一个位置有5个锚框，那么整张图就只有5个尺寸的框，所以不需要roi pooling。

### two stage
主要是fast-rcnn，faster-rcnn这种检测算法。他流程是先提取图像特征，然后生成锚框，再用RPN去进一步筛选锚框得到preposal框。但是需要注意的是preposal的大小是不统一的，所以要用ROI Polling去统一大小，用pooling的原因就是提取出最大特征的前提下统一大小。

之后再对锚框做roi polling缩放到固定的尺寸
